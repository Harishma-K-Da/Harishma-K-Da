Unemployment rate over time

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data
dates = pd.date_range(start='2023-01-01', periods=36, freq='M')
unemployment_rates = np.random.uniform(low=3.0, high=8.0, size=36)

# Create a DataFrame
df = pd.DataFrame({'Date': dates, 'Unemployment_Rate': unemployment_rates})

# Basic Data Analysis
print(df.head())
print(df.describe())

# Data Visualization
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Unemployment_Rate'])
plt.xlabel('Date')
plt.ylabel('Unemployment Rate')
plt.title('Unemployment Rate Over Time')
plt.show()

# More Advanced Analysis (e.g., Time Series Forecasting)
# !pip install statsmodels --upgrade   # If you don't already have statsmodels installed, uncomment and run this
from statsmodels.tsa.arima.model import ARIMA

# Assuming 'Date' is a datetime index
# For this example, p, d, q are set to (5, 1, 0), adjust as needed for your data
model = ARIMA(df['Unemployment_Rate'], order=(5, 1, 0))
model_fit = model.fit()

# Forecast future values
forecast = model_fit.forecast(steps=12)
print(forecast)
```

Percentage change in Unemployment after lockdown in India
